```python
import numpy as np
random_vector = np.random.rand(1, 10)
print(random_vector)
```

```python
import matplotlib.pyplot as plt
random_vector = np.random.rand(10)
plt.hist(random_vector, bins=10, edgecolor='black')
plt.title('Histogram of 1x10 Vector of Random Numbers')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
random_numbers = np.random.randn(500)
print(random_numbers)
```

```python
random_numbers = np.random.randn(500)
plt.hist(random_numbers, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Random Numbers from Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
